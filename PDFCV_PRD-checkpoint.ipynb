```python
def extract_text_from_pdf(pdf_path):
    
    resource_manager = PDFResourceManager()
    fake_file_handle = io.StringIO()
    converter = TextConverter(resource_manager, fake_file_handle)
    page_interpreter = PDFPageInterpreter(resource_manager, converter)

    with open(pdf_path, 'rb') as fh:
        for page in PDFPage.get_pages(fh, 
                                      caching = True,
                                      check_extractable = True):
            page_interpreter.process_page(page)

        plain_text = fake_file_handle.getvalue()

    # close open handles
    converter.close()
    fake_file_handle.close()
    
    if plain_text == '\f':
        plain_text = ""
   
    return plain_text
    
```

```python
def extract_pdf(pdf_path):
    with open(pdf_path, 'rb') as fh:
        plain_text = textract.process(fh,encoding = 'ascii', method = 'tesseract')
    if plain_text == '\f':
        plain_text = ""
    return plain_text
```

```python
def check_valid_format(plain_text):
    check = False
    section_keyword = {'skill': ['skill'], 'experience': ['experience'], 'education': ['education'], 'certification': ['certification','certificate']}
    for key, value in section_keyword.items() :
        for word in value:
            skill = re.search(word, plain_text)
            if skill != None:
                check = True
    # print('check',check)
    return check
```

```python
def weight_score(step3_result , doc_jd, plain_text):
    # score #
    weight_score = col_temp.find_one({'_id' : 'weight_score'})
    weight_score.pop('_id')
    weight_score = copy.deepcopy(weight_score)

    # exp score #
    exp_score = weight_score['work_experience']
    #### candidate exp ####
    print('total_exp_month',step3_result["work_experience"]["total_exp_month"])
    total_exp_month = step3_result["work_experience"]["total_exp_month"]
    if total_exp_month > 0 :
        can_year_exp = int(total_exp_month/12)
    else:
        can_year_exp = 0
    #### weight exp ####
    wei_exp_score = doc_jd['weight_score']['work_experience']['weight']
    exp_score['total'] = doc_jd['weight_score']['work_experience']['total']
    li_wei_exp_score = []
    for wei_exp in  wei_exp_score :
        range_exp = range(int(wei_exp['low']),int(wei_exp['high']))
        exp = copy.deepcopy(wei_exp)
        if can_year_exp in range_exp :
            score_exp = wei_exp['percent']
            exp_score["accumulative_score"] = score_exp
        else :
            score_exp = 0
        exp['score'] = score_exp
        li_wei_exp_score.append(exp)
    exp_score['weight'] = li_wei_exp_score
    weight_score['work_experience'] = exp_score

    # edu score #
    edu_score = weight_score['education']

    #### candidate edu ####
    li_edu = step3_result['education']
    print('li_edu',li_edu)
    print('len(li_edu)',len(li_edu))
    can_edu = empty
    try:
        for edu in li_edu :
            if edu['ref_degree'] == 'DGR004' :
                can_edu = 'DGR004'
                break
            elif edu['ref_degree'] == 'DGR003' :
                can_edu = 'DGR003'
                break
            elif edu['ref_degree'] == 'DGR002' :
                can_edu = 'DGR002'
            else:
                can_edu = 'DGR001'

        print('can_edu',can_edu)
    except:
        if li_edu['ref_degree'] == 'DGR002' :
            can_edu = 'DGR002'
        print('can_edu_2',can_edu)
        pass
    #### weight edu ####
    wei_edu_score = copy.deepcopy(doc_jd['weight_score']['education']['weight'])
    edu_score['total'] = doc_jd['weight_score']['education']['total']
    li_wei_edu_score = []
    for degree_score in wei_edu_score :
        edu = copy.deepcopy(degree_score)
        if can_edu == degree_score['ref_degree'] :
            score_degree = degree_score['percent']
            edu_score['accumulative_score'] = score_degree
        else : 
            score_degree = 0
        edu['score'] = score_degree
        li_wei_edu_score.append(edu)

    edu_score['weight'] = li_wei_edu_score
    weight_score['education'] = edu_score

    # hrd score #
    if step3_result['hard_skill'] != empty:
        hrd_score = weight_score['hard_skill']
        #### candidate hrd ####
        li_hrd = plain_text
        li_can_hrd = li_hrd
        #### weight hrd ####
        wei_hrd_score = copy.deepcopy(doc_jd['weight_score']['hard_skill']['weight'])
        hrd_score["total"] = copy.deepcopy(doc_jd['weight_score']['hard_skill']['total'])
        li_hrd_score = []
        hrd_total_score = 0
        for hrd in wei_hrd_score :
            score_hrd = copy.deepcopy(hrd)
            score_hrd['score'] = 0
            for key in hrd['keyword'] :
            #for weight in li_can_hrd :
                skill = re.search(str(' '+key.lower()+' ') , str(li_can_hrd))
                if skill != None :
                    score_hrd['score'] = hrd['percent']   
            if hrd in li_hrd_score :
                break
            li_hrd_score.append(score_hrd)
            hrd_total_score += score_hrd['score']
        hrd_score['weight'] = li_hrd_score 
        hrd_score["accumulative_score"] = hrd_total_score
        weight_score['hard_skill'] = hrd_score
    else:
        step3_result['hard_skill'] = empty
        hrd_total_score = 0

    # soft score #
    if step3_result['soft_skill'] != empty:
        soft_score = weight_score['soft_skill']
        
        #### candidate soft ####
        li_soft = plain_text
        li_can_soft = li_soft
        #### weight soft ####
        wei_soft_score = copy.deepcopy(doc_jd['weight_score']['soft_skill']['weight'])
        soft_score["total"] = copy.deepcopy(doc_jd['weight_score']['soft_skill']['total'])
        li_soft_score = []
        soft_total_score = 0
        for soft in wei_soft_score :
            score_soft = copy.deepcopy(soft)
            score_soft['score'] = 0
            for key in soft['keyword'] :
                skill = re.search(str(' '+key.lower()+' ') , str(li_can_soft))
                if skill != None :
                    score_soft['score'] = soft['percent']
                    
            if soft in li_hrd_score :
                break
            li_soft_score.append(score_soft)
            soft_total_score += score_soft['score']
        soft_score['weight']= li_soft_score
        soft_score["accumulative_score"] = soft_total_score
        weight_score["soft_skill"] = soft_score
    else:
        step3_result['soft_skill'] = empty
        soft_total_score = 0
    # certificate score #
    cer_score = weight_score['certificate']
    #### candidate certificate ####

    #### weight certificate ####
    try:
        wei_cer_score = copy.deepcopy(doc_jd['weight_score']['certificate']['weight'])
        cer_score["total"] = copy.deepcopy(doc_jd['weight_score']['certificate']['total'])
        li_cer_score = []
        cer_total_score = 0
        for cer in wei_cer_score :
            score_cer = copy.deepcopy(cer)
            score_cer['score'] = 0
            for key in cer['keyword'] :
                skill = re.search(key.lower() , li_can_hrd)
                if skill != None :
                    score_cer['score'] = cer['percent']
                    break    
            if cer in li_cer_score :
                break
            li_cer_score.append(score_cer)
            cer_total_score += score_cer['score']
        cer_score['weight']= li_cer_score
        cer_score["accumulative_score"] = cer_total_score
        weight_score["certificate"] = cer_score
    except:
        cer_total_score = 0 

    total_score = exp_score["accumulative_score"] + edu_score['accumulative_score'] + hrd_total_score + soft_total_score + cer_total_score
    weight_score["accumulative_score"] = total_score
    dic_flow["weight_score"] = weight_score
       
    return dic_flow
```
